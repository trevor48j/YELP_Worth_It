```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import .csv file
result_df=pd.read_csv("Resources/result_clean.csv")
```

```python
# Figure 1 - Make a bar graph for to measure the number of restaurants by cuisine

cuisine_saturation=result_df.groupby(["cuisine"]).count()["name"]
cuisine_df=pd.DataFrame({"number":cuisine_saturation})
cuisine_df_2=cuisine_df.sort_values("number", ascending= False)

cuisine_df_2.plot(kind="bar", 
                  figsize=(16,8), 
                  color = "red", 
                  xlabel="Type of cuisines",
                  ylabel="Number of the restaurants",
                  title="Saturation of the restaurants based on cuisine")

# Save figure to the Resources Folder
plt.savefig("Resources/cuisine_saturation.png")

# Display graph to observe and review
plt.show()

cuisine_df_2
```

```python
# OBSERVATIONS ON FIGURE 1

# American cuisine was found to be the most prevalent in West Hollywood with a total of 19/110 establishments.
# We found that West Hollywood has a total of 36 different cuisines. We also noticed that Argentine, Indian and Korean cuisine have very little to no representation.  
```

```python
# OBSERVATIONS ON FIGURE 2

# We found restaurants that specialize in American cuisine dominate West Hollywood, comprising 17% of all restaurants in the area. Followed by Italian (12%), and Mexican (7%).
```

```python
# Figure 3 - Make a bar graph to plot the average price of restaurants by cuisine

price=result_df.groupby(["cuisine"]).mean()["price"].round(0)
price_df=pd.DataFrame({"Average Price":price})
price_df_2=price_df.sort_values("Average Price", ascending= False).dropna()

plt.rcParams["figure.figsize"] = (16,8)

t1 = price_df_2[price_df_2["Average Price"]<2]
t2 = price_df_2[price_df_2["Average Price"]==2]
t3 = price_df_2[price_df_2["Average Price"]>2]

plt.title('Average price range by type of cuisine')
plt.xlabel('Type of cuisine')
plt.ylabel('Number of "$"')
plt.bar(t1.index.values, t1['Average Price'], color='r')
plt.bar(t2.index.values, t2['Average Price'], color='tab:orange')
plt.bar(t3.index.values, t3['Average Price'], color='gold')
plt.xticks(rotation = 90)

# Save figure to the Resources Folder
plt.savefig("Resources/cuisine_averagepricing.png")

# Display graph to observe and review
plt.show()

price_df_2
```

```python
# OBSERVATIONS ON FIGURE 3

# We determined cuisines that fall within the medium price range account for 71% of all cuisines.
# Next was the expensive tier making up 25% of the cuisine landscape.
# Lastly, the cheap price range represented 3% of total cuisines.
```

```python
# Figure 4 - Make a bar graph to measure the average rating by cuisine

price=result_df.groupby(["cuisine"]).mean()["price"].round(0)
rating=result_df.groupby(["cuisine"]).mean()["rating"].round(0)
correlation_df=pd.DataFrame({"Average Price":price,
                    "Average Rating":rating})
correlation_df_2=correlation_df.sort_values("Average Price", ascending= False).dropna()

expensive=correlation_df_2["Average Price"].replace(3.0, "expensive")
correlation_df_2["Average Price"]= expensive

medium=correlation_df_2["Average Price"].replace(2.0, "medium")
correlation_df_2["Average Price"]= medium

cheap=correlation_df_2["Average Price"].replace(1.0, "cheap")
correlation_df_2["Average Price"]= cheap


correlation=correlation_df_2.groupby(["Average Price"]).mean()["Average Rating"]
correlation_df_3 = pd.DataFrame({
                                "Average Rating" : correlation
})

correlation_df_3.plot(kind="bar", 
                      color="r", 
                      xlabel="Price Range", 
                      ylabel="Average Rating", 
                      title="Average Ratings by Average Price Range")
plt.xticks(rotation = 0)

# Save figure to the Resources Folder
plt.savefig("Resources/pricerange_averagerating.png")

# Display graph to observe and review
plt.show()

correlation_df_3
```

```python
# OBSERVATIONS ON FIGURE 4: 

# From our data we've uncovered that cheaper restaurants generated, on average, a 5-star rating. We believe this to be due to lower expectations to cheaper costing food thus allowing room to gain better ratings.
# Our graph showed us that the city of West Hollywood has generally high ratings across the different price ranges and could be attributed to the critical food culture in that neighborhood.
```

```python
# Figure 5 - Make a bar graph to measure the number of restaurants by average price range

comparison=correlation_df_2.reset_index()
comparison_data = comparison.groupby(["Average Price"]).count()["cuisine"]

comparison_df = pd.DataFrame({"number of restaurants":comparison_data})

plt.title('The number of Restaurants by average price range')
plt.xlabel('Average Price Range')
plt.ylabel('Number of Restaurants')
t1 = comparison_df[comparison_df["number of restaurants"]<4]
t2 = comparison_df[comparison_df["number of restaurants"]<11]
t3 = comparison_df[comparison_df["number of restaurants"]>12]
plt.bar(t1.index.values, t1['number of restaurants'], color='tab:orange')
plt.bar(t2.index.values, t2['number of restaurants'], color='gold', alpha= 0.5)
plt.bar(t3.index.values, t3['number of restaurants'], color='r')

# Save figure to the Resources Folder
plt.savefig("Resources/averageprice_numberofrestaurants.png")

# Display graph to observe and review
plt.show()

comparison_df
```

```python
# OBSERVATIONS ON FIGURE 5

# rom our analysis, 65% of restaurants fall within the medium price range, with Expensive (32%) and cheap (3%). 
# Generally speaking, we believe that medium tiered restaurants represent the best of both the cheap and expensive. 
# Straddling the middle will allow customers to experience the quality and service that expensive restaurants have to offer. 
# While offering a pricing structure that is more accessible to a wider audience.
```
